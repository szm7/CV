#Write & Convert

```python
import imageio as iio

img = iio.imread("k1.jpg")
iio.imwrite("k1p.png", img)
iio.imwrite("k1j.jpeg", img)
iio.imwrite("k1j.jpe", img)
iio.imwrite("k1j.tiff", img)
iio.imwrite("k1j.tif", img)
```
