```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread(r'/content/k4.jpg',0)
```

```python
plt.imshow(image, cmap='gray')
plt.title('Original Imported image')
plt.show()
```

**Image Negative**

```python
img_negative1 = cv2.bitwise_not(image)
```

```python
img_negative2 = 255 - image
```

```python

plt.figure(figsize=(10, 5))
plt.subplot(132), plt.imshow(img_negative1, cmap='gray'), plt.title('Built in')
plt.subplot(133), plt.imshow(img_negative2, cmap='gray'), plt.title('User defined')

plt.show()
```

```python
def gray_level_slicing_preserve(image, lower_bound, upper_bound):
    result_image = image.copy()
    result_image[(image >= lower_bound) & (image <= upper_bound)] = 255
    return result_image

sliced_image = gray_level_slicing_preserve(image, 100, 150)

plt.imshow(sliced_image, cmap='gray')
plt.title('Gray Level Slicing')
plt.show()
```

```python
equalized_image = cv2.equalizeHist(image)

plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(equalized_image, cmap='gray'), plt.title('Equalized Image')
plt.show()
```

```python
def histogram_equalization(image):
    histogram, bins = np.histogram(image.flatten(), bins=256, range=[0,256])
    cdf = histogram.cumsum()
    cdf_normalized = cdf * float(histogram.max()) / cdf.max()
    image_equalized = np.interp(image.flatten(), bins[:-1], cdf_normalized)
    return image_equalized.reshape(image.shape)

equalized_image = histogram_equalization(image)
```

```python

plt.figure(figsize=(10, 5))
plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(equalized_image, cmap='gray'), plt.title('Equalized Image')
plt.show()
```
