```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy import ndimage
from scipy.signal import convolve2d
from scipy.ndimage import gaussian_filter
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
from scipy.ndimage import gaussian_filter
```

reading image using cv2 library

```python
image = cv2.imread( "/content/k1.jpg", 0)
```

defining different kernal matrix for image operations

```python
kernel_identity = np.array([[0, 0, 0], [0, 1, 0], [0, 0, 0]])

kernel_edge = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])

kernel_sharpen = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])

kernel_blur = (1 / 9.0) * np.array([[1, 1, 1], [1, 1, 1], [1, 1, 1]])

kernel_gaussian = (1 / 256.0) * np.array([
    [1, 4, 6, 4, 1],
    [4, 16, 24, 16, 4],
    [6, 24, 36, 24, 6],
    [4, 16, 24, 16, 4],
    [1, 4, 6, 4, 1],
])
```

processing the image with different kernals

```python
dst_identity = cv2.filter2D(image, -1, kernel_identity)
dst_edge = cv2.filter2D(image, -1, kernel_edge)
dst_sharpen = cv2.filter2D(image, -1, kernel_sharpen)
box_blur_result = convolve2d(image, kernel_blur)
gaussian_blur_result = gaussian_filter(image, sigma=1)
```

plotting image using matplotlib library

```python
plt.figure(figsize=(20, 15))

plt.subplot(6, 1, 1), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(6, 1, 2), plt.imshow(dst_identity, cmap='gray'), plt.title('Identity Kernel')
plt.subplot(6, 1, 3), plt.imshow(dst_edge, cmap='gray'), plt.title('Edge Detection Kernel')
plt.subplot(6, 1, 4), plt.imshow(dst_sharpen, cmap='gray'), plt.title('Sharpen Kernel')
plt.subplot(6, 1, 5), plt.imshow(box_blur_result, cmap='gray'), plt.title('Box Blur Result')
plt.subplot(6, 1, 6), plt.imshow(gaussian_blur_result, cmap='gray'), plt.title('Gaussian Blur Result')

plt.tight_layout()
plt.show()
```

convolve function with different modes

```python

kernel = np.array([[1, 1, 1],
                   [1, 0, 1],
                   [1, 1, 1]])

result_reflect = convolve(image, kernel, mode='reflect')

result_constant = convolve(image, kernel, mode='constant', cval=0)

result_nearest = convolve(image, kernel, mode='nearest')

result_mirror = convolve(image, kernel, mode='mirror')

result_wrap = convolve(image, kernel, mode='wrap')

plt.figure(figsize=(10, 5))

plt.subplot(2, 3, 1), plt.imshow(image, cmap='gray'), plt.title('Original')
plt.subplot(2, 3, 2), plt.imshow(result_reflect, cmap='gray'), plt.title('Reflect')
plt.subplot(2, 3, 3), plt.imshow(result_constant, cmap='gray'), plt.title('Constant')
plt.subplot(2, 3, 4), plt.imshow(result_nearest, cmap='gray'), plt.title('Nearest')
plt.subplot(2, 3, 5), plt.imshow(result_mirror, cmap='gray'), plt.title('Mirror')
plt.subplot(2, 3, 6), plt.imshow(result_wrap, cmap='gray'), plt.title('Wrap')

plt.show()
```
